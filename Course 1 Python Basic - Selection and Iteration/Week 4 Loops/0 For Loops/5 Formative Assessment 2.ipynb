Which of the following code snippets will loop through all number between 0 and 100 (including 100)?

- range(100)
- range(101)  ---> Correct Answer
- range(0, 100)
- range(0, 101, 1) ---> Correct Answer

```python
for i in range(101):
    print(i)
```

```python
for i in range(0,101,1):
    print(i)
```

    Remember, the range function does not include the ending number. So range(100) does not include 100. range(0, 101, 1) is correct because this means you increment from 0 to 100 by 1.
