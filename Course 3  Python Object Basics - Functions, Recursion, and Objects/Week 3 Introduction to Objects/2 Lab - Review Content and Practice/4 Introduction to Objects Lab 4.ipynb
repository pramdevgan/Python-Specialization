### Add Caption

The font in the caption label will be similar to the font for the username, except it will not be bold. The `wraplength`parameter determines when the word wrap begins. The bigger the number, the more you can write before word wrap happens. The space to the right of the main photo is actually two columns. We want the caption to fill this entire area, so the text needs to span across two columns.

```python
# User caption for the photo
caption = tkinter.Label(
    window,
    text=post1.caption,
    font="DejaVuSans 14",
    wraplength=300,
    justify="left",
    bg="white")

caption.grid(
    row=1,
    column=1,
    columnspan=2,
    sticky="NW")
```

````python
class Post:
    """Create a post object for the fictitious social network"""
    def __init__ (self, username, user_id, media, avatar, comment_button, caption, likes, comments, like_button):
        self.username = username
        self.user_id = user_id
        self.media = media
        self.avatar = avatar
        self.comment_button = comment_button
        self.caption = caption
        self.likes = likes
        self.comments = comments
        self.like_button = like_button

username = "Sally_17"
user_id = 112010
media = "student_folder/img/photogram/waterfall.png"
avatar = "student_folder/img/photogram/avatar_icon.png"
comment_button = "student_folder/img/photogram/add_comment.png"
caption = "First time at Yosemite. It has surpassed all of my expectations."
likes = 23
comments = ["Beautiful!", "I wish I was there too.", "Is that Nevada Falls?", "Love it!", "Can't wait for the Halfdome pictures", "More pics please"]
like_button = "student_folder/img/photogram/likes_icon.png"

post1 = Post(username, user_id, media, avatar, comment_button, caption, likes, comments, like_button)
print(post1.username, post1.user_id, post1.media, post1.avatar, post1.comment_button,post1.caption, post1.likes, post1.comments, post1.like_button)
````

````python
import tkinter
window = tkinter.Tk()
window.title("Photogram")
window.geometry("800x500")
window.configure(background="white")
window.grid_columnconfigure(1, weight=0)
window.grid_columnconfigure(2, weight=1)
photo = tkinter.PhotoImage(file=post1.media)
comment_button = tkinter.PhotoImage(file=post1.comment_button)
avatar = tkinter.PhotoImage(file=post1.avatar)
like_button = tkinter.PhotoImage(file=post1.like_button)
# Big photo on the left
image = tkinter.Label(
    window,
    image=photo,
    bg="white")

image.grid(
    row=0,
    column=0,
    rowspan=10,
    stick="W")

# Gray user avatar
user_avatar = tkinter.Label(
    window,
    image=avatar,
    width=30,
    bg="white")

user_avatar.grid(
    row=0,
    column=1,
    sticky="W")
# Username to the right of the avatar
user_name = tkinter.Label(
    window,
    text=post1.username,
    font="DejaVuSans 14 bold",
    justify="left",
    bg="white")

user_name.grid(
    row=0,
    column=2,
    sticky="W")

# User caption for the photo
caption = tkinter.Label(
    window,
    text=post1.caption,
    font="DejaVuSans 14",
    wraplength=300,
    justify="left",
    bg="white")

caption.grid(
    row=1,
    column=1,
    columnspan=2,
    sticky="NW")

# Add comment icon
comment_icon = tkinter.Label(
    window,
    image=comment_button,
    bg="white",
    justify="center")

comment_icon.grid(
    row=2,
    column=1,
    columnspan=2)
window.mainloop()  # This should be the last line of code in your program
````

## Lab Question

what is a default parameter?

- A default parameter is required when using the constructor.
- A default parameter is another name for an object.
- ***A default parameter is a parameter for the constructor that has a default value. Users can override this default instantiating an object.*** <--- Correct Answer
- The default parameter for the constructor is `self`.


A default parameter is a parameter for the constructor that has a default value. That means when a user instantiates an object, they can omit this parameter and constructor will use the default value. Or, they can pass the constructor a value which will override the default value. In the code sample below, `param2` is an example of a parameter with the default value of 10.

```python
class ExampleClass:
  def __init__(self, param1, param2=10):
    self.param1 = param1
    self.param2 = param2

example1 = ExampleClass(5, 20) # param1 = 5, param2 = 20
example2 = ExampleClass(7)# param1 = 7, param2 = 10
```
